```python
def calc(a,b,c):
    if(a.isnumeric() == True) and (c.isnumeric() == True):
        a = int(a)
        c = int(c)
        if b == 'x':
            print('Hasil dari', a, b, c, 'adalah', a*c)
        elif b == ':':
            print('Hasil dari', a, b, c, 'adalah', a/c)
        elif b == '+':
            print('Hasil dari', a, b, c, 'adalah', a+c)
        elif b == '-':
            print('Hasil dari', a, b, c, 'adalah', a-c)
        elif b == '^':
            print('Hasil dari', a, b, c, 'adalah', a**c)
        else:
            print('operator', b, 'tidak ditemukan')
    else:
        print('salah')

a = input('masukkan angka1= ')
b = input('masukkan operator= ')
c = input('masukkan angka2= ')

calc(a,b,c)
```

### Quiz 2

```python
def kelompok_3(tinggi):
    if tinggi%2 == 0:
        tinggi -= 1
    tengah = int(tinggi/2) + 1
    z = ''
    for i in range(tengah):
        z += '  '*i+'* '*(tinggi-i*2)+'\n'
    for i in range(tengah-1):  
        j = tengah-1 - (i+1)
        z += '  '*j+'* '*(tinggi-j*2)+'\n'
    return z
a = int(input('how tall? '))
print(kelompok_3(a))
```

## Prime Number Generator Based on Interval

```python
# First, we will take the input:
lower_value = int(input('Please, enter the lowest range value: '))
upper_value = int(input('Please, enter the upper range value: '))

print('The prime numbers in the range are: ')
for number in range (lower_value, upper_value + 1):
    if number > 1:
        for i in range(2, number):
            if (number % i) == 0:
                break
        else:
            print(number)
```

```python
def generate_primes(lower, upper):
    primes = []
    for num in range(lower, upper + 1):
        # bilangan prima harus lebih besar dari 1
        if num > 1:
            # cek apakah bilangan tersebut habis dibagi dengan bilangan lain
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                primes.append(num)
    return primes

# contoh penggunaan
lower = int(input("Masukkan nilai terendah: "))
upper = int(input("Masukkan nilai tertinggi: "))

print(f"Bilangan prima antara {lower} dan {upper}:")
print(generate_primes(lower, upper))
```

```python
def print_odd_numbers(start, end):
   
    if start % 2 == 0:
        start += 1
          
    for i in range(start, end+1, 2):
        print(i)
a = int(input("Masukkan angka terendah: "))
b = int(input("Masukkan angka tertinggi: "))
print_odd_numbers(a, b)
```
