## If Elif and Else

```python
harga_jeruk = 20000
standar_harga = 30000
# apabila harga jeruk lebih dari standar harga maka jeruk itu mahal, apabila harga jeruk kurang dari standar harga maka jeruk itu murah, apabila harga jeruk sama dengan standar harga maka biasa
if harga_jeruk > standar_harga:
    print('Mahal')
elif harga_jeruk < standar_harga:
    print('Murah')
else:
    print('Biasa')
```

### Quiz 2
Anda adalah seorang pemilik perusahaan dan Anda berniat untuk memberi bonus kepada pegawai Anda.
Buatlah program if-else yg memiliki kondisi, jika Year of Service seorang pegawai lebih dari 10 tahun, maka gajinya akan diberi bonus sebesar 10%.
contoh:
* in: berapa gaji anda? 10000000
* in: year of service? 15
* out: Selamat Anda mendapat bonus! Total gaji anda IDR 11000000

```python
gaji = int(input('Berapa gaji Anda? '))
year_of_service = int(input('Year of Service? '))

if year_of_service > 10:
    bonus = 0.1

    total_gaji = gaji + (bonus * gaji)

    print('Selamat Anda mendapat bonus! Total gaji Anda IDR', total_gaji)

else:
    print('Maaf, Anda tidak mendapatkan bonus')
```

## While Loop

```python
apel = 5
while apel > 0:
    apel -= 1
    print(apel)
```

```python
apel = 5
apel -= 1
print(apel)
```

```python
apel = 5
while apel > 0:
    apel -= 1
    if apel == 1:
        print('tinggal 1 nih')
    else:
        print(apel)
```

```python
# break digunakan untuk memberhentikan loop
apel = 5
while apel > 0:
    apel -= 1
    if apel == 2:
        break
    print(apel)
print('Loop ended')
```

```python
apel
```

```python
# continue digunakan untuk melanjutkan loop berdasarkan kondisi
apel = 5
while apel > 0:
    apel -= 1
    if apel == 2:
        continue
    print(apel)
print('Loop ended')
```

```python
# else
apel = 10
while apel > 0:
    apel -= 2
    print(apel)
else:
    print('Loop ended')
```

```python
# break digunakan untuk memberhentikan loop
apel = 5
while apel > 0:
    apel -= 1
    print(apel)
    if apel == 2:
        break
else:
    print('Loop ended')
```

```python
a = [1,2,3,4,5,6]
for i in a:
    print(i)
```

```python
b = ['Budi', 'Joko', 'Anwar']
for i in b:
    print(i)
```

```python
# Nested for loop
b = ['Budi', 'Joko', 'Anwar']
for i in b:
    for j in i:
        print(j)
```

```python
# Nested for loop
b = ['Budi', 'Joko', 'Anwar']
for i in b:
    if i == 'Joko':
        break
    for j in i:
        print(j)
```

```python
# Continue
for i in ['Aster', 'Jubah', 'Nambo', 'Hachi']:
    if 'b' in i:
        continue
    print(i)
```

```python
# Pass
for i in ['Aster', 'Jubah', 'Nambo', 'Hachi']:
    if 'b' in i:
        pass
    print(i)
```

```python
# Break
for i in ['Aster', 'Jubah', 'Nambo', 'Hachi']:
    if 'b' in i:
        break
    print(i)
```

* Loop itu keseluruhan iterasi objek 1 loop diatas 4 iterasi
* Iterasi adalah istilah untuk 1 putaran loop
* continue akan memberhentikan iterasi dan lanjut ke iterasi berikutnya
* break akan memberhentikan looping secara keseluruhan

```python
for i in ['Aster', 'Jubah', 'Nambo', 'Hachi']:
    print(i)
else:
    print('Loop Done')
```

```python
# Break
for i in ['Aster', 'Jubah', 'Nambo', 'Hachi']:
    if 'b' in i:
        break
    print(i)
else:
    print('Loop Done')
```

### Quiz 3
Menggunakan while loop buatlah program sederhana.
ketika bertemu dengan angka ganjil, maka print angkanya.
ketika bertemu dengan angka genap, maka print "GENAP"
batas angka 1-20

```python
n = 1
while n <= 20:
    if n % 2 == 1:
        print(n)
    else:
        print('genap')
    n += 1
```

```python
n = int(input())
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=' ')
    print('')
```
